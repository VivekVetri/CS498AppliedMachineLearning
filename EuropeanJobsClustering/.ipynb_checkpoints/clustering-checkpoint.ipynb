## Single Linkage

```python
# Part of the code is obtained from https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html#scipy.cluster.hierarchy.linkage
Z = linkage(data.iloc[ : , 1:len(data.columns)], 'single')
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z, labels=list(data.iloc[:,0]))
plt.title("Single Linkage")
plt.show()
```

```python
Z = linkage(data.iloc[ : , 1:len(data.columns)], 'complete')
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z, labels=list(data.iloc[:,0]))
plt.title("Complete Linkage")

plt.show()
```

```python
Z = linkage(data.iloc[ : , 1:len(data.columns)], 'average')
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z, labels=list(data.iloc[:,0]))
plt.title("Average Linkage")
plt.show()
```

```python
inertia=[]
for c in range(2,26):
    kmeans = KMeans(n_clusters= c, n_init=1, random_state=1)
    kmeans.fit(data.iloc[ : , 1:len(data.columns)])
    print("Inertia using 2 clusters = {}".format( kmeans.inertia_))
    inertia.append(kmeans.inertia_)

```

```python
inertiadf = pd.DataFrame({"clusters":range(2,26), "error":inertia})
plt.plot(inertiadf.error)
plt.xticks(range(0,24),inertiadf.clusters)
plt.xlabel("Clusters")
plt.ylabel("Inertia")
plt.show()
```

#### Using elbow method and by visualizing elbow plot above, we can select number of clusters to be 5. Our goal here is to find lowest numbers of cluster that also has good inertia value. Interia has been drastically decresed by using 5 cluster and after that decreate rate of inertia is very small. In addition, this dataset doesnt form a clear elbow. This might be because the size of our dataset is really small and this small dataset might not be very clustered.
